```
import sys
import ast

def parse_notebook_params() -> dict:
    args = {}
    if '' in sys.argv:
        sys.argv.remove('')    # remove empty strings
    if len(sys.argv):          # do nothing if no param supplied
        for arg in sys.argv:   # sys.argv evaluates to a list 
            k, v = arg.split('=')
            v = ast.literal_eval(v)
            args[k] = v
    return args

default_args = dict(
    wait_seconds=5,
    raise_error=False,
)

args = parse_notebook_params()

for key in default_args.keys():    # look for default values
    if key not in args:            # if not supplied then assign
        args[key] = default_args[key]

print(args)
```

```
SELECT SYSTEM$WAIT({{args['wait_seconds']}})
```

```
def force_stop(raise_error: bool):
    if raise_error:
        raise RuntimeError("An error was raised because the input is True.")
    else:
        pass

force_stop(args['raise_error'])
```
