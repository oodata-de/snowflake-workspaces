```
from snowflake.snowpark.context import get_active_session
session = get_active_session()

query_list = [
    "EXECUTE NOTEBOOK DUMMY_NOTEBOOK_TEMPLATE('wait_seconds=30 raise_error=False')"
  , "EXECUTE NOTEBOOK DUMMY_NOTEBOOK_TEMPLATE('wait_seconds=30 raise_error=False')"
  , "EXECUTE NOTEBOOK DUMMY_NOTEBOOK_TEMPLATE('wait_seconds=30 raise_error=False')"
]

import time
t0 = time.perf_counter()

job_list = []
for query in query_list:
    query_id = session.sql(query).collect_nowait().query_id
    async_job = session.create_async_job(query_id)
    job_list.append(async_job)

while len(job_list):
    for job in job_list:
        if job.is_done():
            df = job.result()
            print(df)
            job_list.remove(job)

t1 = time.perf_counter()
time_used = t1 - t0
print(f"All queries finished after {time_used:.2f} seconds")
```
